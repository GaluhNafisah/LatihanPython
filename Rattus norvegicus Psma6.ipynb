```python
import os
#creating EFetch handle
filename = "Rattus norvegicus Psma6.fas"

#creating Efetch handle
if not os.path.isfile(filename):
    #Downloading file
    efetch_handle = Entrez.efetch(db="nuccore" , id="402747073" , rettype= "fasta", retmode= "text")
    output_file_handle = open(filename, "w")
    output_file_handle.write(efetch_handle.read())
    output_file_handle.close()
    efetch_handle.close()
    print("FASTA Download")
```

```python
import os
#creating EFetch handle
filename = "Rattus norvegicus Psma6.gbk"

#creating Efetch handle
if not os.path.isfile(filename):
    #Downloading file
    efetch_handle = Entrez.efetch(db="nuccore" , id="402747073" , rettype= "gb", retmode= "text")
    output_file_handle = open(filename, "w")
    output_file_handle.write(efetch_handle.read())
    output_file_handle.close()
    efetch_handle.close()
    print("GB Downloaded")
```

```python
from Bio import SeqIO

GoI = "Psma6"
for genome in SeqIO.parse("Rattus norvegicus Psma6.gbk","genbank"):
    for gene in genome.features:
        if gene.type != "CDS":
            continue
        if "gene" not in gene.qualifiers["gene"][0]:
                continue
        if GoI not in gene.qualifiers["gene"][0]:
                    continue
    gene_seq = gene.extract(genome.seq)
                    
GoI_fasta = ">%s|%s\n%s" % (GoI, genome.description, gene_seq)
print(GoI_fasta)
```

```python
from Bio import SeqIO

GoI = "Psma6"
for genome in SeqIO.parse("Rattus norvegicus Psma6.gbk","genbank"):
    for gene in genome.features:
        if gene.type != "CDS":
            continue
        if "gene" not in gene.qualifiers["gene"][0]:
                continue
        if GoI not in gene.qualifiers["gene"][0]:
                    continue
    gene_seq = gene.extract(genome.seq)
                    
GoI_fasta = ">%s|%s\n%s" % (GoI, genome.description, gene_seq)
with open("GoI_fasta.fasta","w") as f:
    f.write(GoI_fasta)
```
